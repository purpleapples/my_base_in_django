# 구로시설 관리공단 경영평가지표 데이터 정의
## 사용자
- 해당 시스템을 이용하는 대상으로 기획 예산 팀 이외의 각 사업들을 관장하는 사업장의 인원이 이용
## 사용자 정보
- 해당 시스템을 이용하는 대상이 가지고 있는 부가 정보 이 중 계정 종류 코드는 권한 통제용으로 사용
## 메뉴
- 사용자가 접근을 메뉴에 대한 정보를 계층구조로 관리
## 메뉴 권한
- 사용자가 접근 할 수 있는 메뉴에 대한 권한을 관리. 팀, 계정, 계정 종류 코드로 관리
## 화면
- 메뉴에 연동되는 화면에 대한 정보와 화면의 접근 주소 관리
## 화면 권한(미사용)
- 화면에서 작동되는 기능에 대한 권한 관리
## 게시글
- 게시글 관리
## 게시글 관계
- 게시글 중 질문 답변 관계가 형성되는 게시물에 대한 관리
## 팀
- 각 사업장을 관리하는 인원이 속한 집단을 지칭, 예산, 인건비, 시설 관련 원천데이터의 기준값으로 사업이 있으며, 사업 관련 정보는 팀이 등록 관리한다.
## 사업
- 구로시설 관리공단에서 운영하는 사업을 지칭하며 하나에서 다수의 사업장을 사업장 중 대표로 선장된 팀이 관리한다.
## 사업 시설
- 각 사업에 종속된 시설. 관리면적이 존재한다.
## 사업 시설 운영 프로그램
- 각 사업 시설에서 진행되는 프로그램. 각 사업은 월별 프로그램 이용자 수, 수임금을 입력관리한다.
## 회계 입력 기간
- 각 사업장에서는 각 사업장의 전월 프로그램의 이용자수, 수익금에 대해 당월 10일 까지 입력을 허용한다.
- 허나 이 이외의 기간에 입력, 수정을 실시하는 경우 회계입력기간이 열려있어야 가능하다.
## 평가 방법 관리
- 각 지표에 쓰이는 평가 방법의 타입과 방법 명칭, 최고 목표 계산 가중치, 최저 목표 계산 가중치, 가중치, 득점, 평점의 계산 방법을 서술한다.
## 코드 테이블
- 시스템의 기반 데이터는 모두 이곳에 저장된다.
- 계층구조를 이루고 있으며 코드값은 unique 해야 한다.
## 연별 코드 사용
- 연도별로 사용할 예산 과목 코드, 인건비 상세 항목 코드에 대해 지정, 관리한다.
## 정규직 인건비 인상율
- 연년별 정규직의 인건비 인상율을 관리한다.
## 조정 코드 상세
- 회계 내역 조정을 위한 코드가 가지는 조정 범위 코드를 설정, 관리
## 조정 항목 코드
- 조정 시 조정 대상이 될 항목에 들어갈 코드를 관리
## 코드 그룹
- 계층 구조가 아닌 서로의 용도에 의해 묶일 코드들을 관리 - 최저임금 인상분 계산을 위한 인건비 그룹, 노동생산성을 위한 복리후생과목, 세금과공과과목에 들어갈 항목을 정의 시 사용
## 예산 사용 기록
- 사업별, 월별, 예산과목별 예산액과 지출액에 대해 저장
## 인건비
- 사업별, 직원별, 월, 지급일자 별 인건비 세부항목을 저장. 인건비 세부항목의 경우 JSONFIELD로 해당 정보를 한줄에 저장
## 외부인건비
- 사업별 월별 외부인원(시간강사 등)의 인건비를 저장
## 사업프로그램월별이용기록
- 사업 시설 운영 프로그램의 년도별 월별 이용실적과 수익금을 관리
## 연도별 월초 월말 임직원 통계
- 연도별로 월초, 월말의 직급별 근무인원, 조정 사유별 제외인원 관리
## 조정
- 예산, 시설 프로그램기록, 인건비 에 대해 제목, 조정기간, 조정내역을 관리
- 모든 조정 내역은 한달을 30일로 보고 금액 인원에 대해서 해당월의 수치/ 해당월의 일자 * 30 으로 치환한 뒤 진행한다.
## 예산사용기록 조정
- 조정을 참조하며 조정 대상 범위가 사업, 예산 과목 중 어떤 것을 활용했는지를 관리
- 사업 조정 시 사업의 모든 예산과목이 조정대상이 된다.
- 예산 조정 시 모든 사업의 해당 기간 동안의 예산
- 사업 - 예산 을 조정 시 해당 사업 의 해당 예산 과목이 조정 대상이 된다.
## 인건비 조정
- 조정을 참조하며 조정 대상 범위가 사업, 직원, 인건비 항목 중 어떤 것을 활용했는지를 관리
- 사업, 직원, 인건비는 다중 선택을 허용하며 다중 선택 시 해당 내역에 걸리는 것은 모두 조정 대상이 된다.
## 시설프로그램 조정
- 조정을 참조하며 조정 대상 범위가 사업, 시설, 프로그램 중 어떤 것을 활용했는지를 관리
- 사업, 시설, 프로그램은 다중 선택을 허용하며 다중 선택 시 해당 내역에 걸리는 것은 모두 조정 대상이 된다.
- 프로그램 조정 시 시설 하나가 조정되었다고 볼 수 있는 경우 관리면적을 조정한다.
## 조정 계산외 기록
- 해당 시스템에 서 계산 할 수 없는 내역을 연간 단위로 사업 혹은 주제 별로 조정액 관리

# 구로시설관리공단 데이터 flow
## 0. 필수 데이터 설정
- 계정 생성 -> 팀 생성 -> 사업 생성 -> 팀 인원 설정
- 대행사업 절감률, 사업수익, 인건비, 노동생산성 평가 방법 입력
## 1. 대행 사업 절감률
- 2년치 예산 사용기록 업로드 -> 인건비 사용 항목 관리에서 연도별로 사용할 예산과목 설정 -> 예실대비목록 확인
- 예실대비 항목 조정 -> 대행 사업 절감률 표기용 예산과목 설정 -> 최저 임금 인상분 계산용 항목 설정 -> 대행사업절감률 확인
## 2. 결산내역 기준 인건비
- 예산 사용 기록 업로드 (인건비 항목에 귀속 및 결산내역 등록용) -> 2년치 인건비 업로드 -> 인건비 항목 조정 -> 연도별 인건비 산출 확인
## 3. 사업 수익
- 시설 이용 현황 에서 연도별 월별 사업 이용인원, 수익금 등록 -> 시설 조정 등록 -> 사업수익
## 4. 노동생산성
- 1,2,3 완료. -> 복리후예산과목 등록, 세금과 공과 예속 예산과목 등록 -> 영업이자 등록 ->  노동생산성 지표 확인

# 각 지표별 사용 데이터 산출 방식
## 1. 대행사업비 절감률
- 영업비용의 지출액 을 결산액으로 지칭
- 해당 결산액에서 사업별 조정내역 계산
- 예산 단독 조정 과목 존재 여부 확인
    - 존재 시 사업별 조정액에서 해당 과목의 조정 금액 차감
    - 예산 단독 조정 과목의 조정액을 사업 기록 하단에 부착
- 인건비 그룹에 존재하는 내역을 예산 과목의 지출 액 합산에 인건비 인상률 곱해서 최저 임금 인상분 * 하고 합산
- 조정 내역 수기입력 부착
- 물가상승률 입력
- 전년 결산액 - 조정액 의 합계를 대상으로 최고/최저 목표 산출 후 가중치, 평점, 득점 계산
-
## 2. 결산내역 기준 인건비
- 예산 사용 기록의 영업비용의 지출액을 결산액으로 지칭
- 2년치 사업별 인건비 불러오기
- 2년치 사업별 인건비에 조정 액 계산
- 2년치 외부인건비 불러오기

## 3. 사업 수익
- 2년치 시설 운영 프로그램 데이터 불러오기
- 조정 기록을 대상으로 조정액 계산
- 사업별 운영프로그램 수입 합산
- 조정액을 사업별 합산

## 4. 노동생산성
- 필요 데이터 : 사업 프로그램 월간 이용 기록, 인건비, 연간 임직원 월초 월말 인원 통계, 월별예산사용기록
- 산출 테이블 : 작성양식, 연간 임직원 월초 월말 인원 테이블, 연도별 세부 조정 내역
- 산출 테이블 별 세부 산출 방식
    - 작성 양식
        - 사업 수지 :
            - 사업 수입: 사업 프로그램 월간 이용 기록을 참조하되 공익목적사업은 배제
            - 대행사업비용 : 손익계산서의 판매관리비 + 매출원가 + 영업외비용 (지방 공기업 최고경영자협의체부담금) -> 예산 내역의 영업이익
        - 인건비 :
            - 매출 원가의 인건비 : 예산 과목 그룹의 매출 원가의 인건비
            - 매출원가의 복리후생비 : 예산 과목 그룹의 매출 원가의 복리후생비
            - 판관비의 인건비 : 예산 과목 그룹의 매출 원가의 인건비
            - 판관비의 복리후생비 : 예산 과목 그룹의 판매 관리비의 복리후생비
        - 수입이자 : 영업 이자
        - 세금과공과 : 예산 과목 그룹의 세금과 공과
        - 감가상각비 : 예산과목
            - 매출원가의 감가상각비 : 예산 과목의 감가상각비
            - 판관비의 감가상각비 : 예산 과목의 감가상각비(판)
    - 연별 월말 월초 임직원 인원 평균
        - 포함인원 - 제외 인원 내역
    - 연도별 세부조정내역
        - 사업수지
            - 사업수익
               - 사업프로그램월간이용기록에 대한 사업기록조정내역
            - 대행사업비용 : 작성양식의 조정비용
        - 인건비
            - 매출원가의 인건비 : 행정인턴 어떻게구하는지?
            - 매출원가의 복리후생비 : 예산사용기록에 대한 사업별 조정액 및 기록
        - 수입이자 : 영업이익
        - 세금과공과 : 작성양식의 세금과공과 중 조정액
        - 감가상각비
            - 매출원가의 감가상각비 : 작성양식의 매출원가의 감가상각비
            - 판관비의 감가상각비 : 작성양식의 판관비의 감가상각비
